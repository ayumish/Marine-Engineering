# Model Building for target 1

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Splitting the data into training and testing sets for target1
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_scaled, y1, test_size=0.2, random_state=42)

# Splitting the data into training and testing sets for target2
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_scaled, y2, test_size=0.2, random_state=42)


# Initialize the RandomForestRegressor for target1
rf_model1 = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model for target1
rf_model1.fit(X_train1, y_train1)

# Make predictions for target1
y_pred1 = rf_model1.predict(X_test1)

```
